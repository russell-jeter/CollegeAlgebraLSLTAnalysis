```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def item_c_curve_func(theta, delta):
    return np.exp(theta - delta) / (1 + np.exp(theta - delta))
```

```python
theta_list = np.linspace(-4, 4, 1000, endpoint=True)
delta_list = np.round(np.linspace(-1, 1, 5, endpoint=True), 2)
```

```python
func, ax = plt.subplots(1, 1)

for delta in delta_list:
    plt.plot(theta_list, item_c_curve_func(theta_list, delta))

custom_palette = ['#cf4456', '#f29566', '#831c64', '#2f0f3e', '#feedb0']
plt.rcParams['axes.prop_cycle'] = plt.cycler('color', custom_palette)
ax.legend(delta_list, prop={'size': 10})
ax.set_xlabel("ability")
ax.set_ylabel("probability correct")
ax.set_title("Item Characteristic Curve")

plt.savefig('./figures/item_characteristic_curve_example.png')
```
