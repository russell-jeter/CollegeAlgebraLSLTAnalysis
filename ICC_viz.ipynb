```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sb
```

```python
def item_c_curve_func(theta, delta):
    return np.exp(theta - delta) / (1 + np.exp(theta - delta))
```

```python
theta = np.linspace(-4, 4, 1000, endpoint=True)
delta = np.round(np.linspace(-1, 1, 5, endpoint=True), 2)

icc_data_df = pd.DataFrame(theta, columns=["ability"])

for d_temp in delta:
    icc_data_df[d_temp] = item_c_curve_func(theta, d_temp)

final_df = pd.melt(icc_data_df, id_vars='ability', var_name='delta', value_name='percent')
final_df.set_index('ability', inplace=True)
icc_data_df.set_index('ability', inplace=True)
```
